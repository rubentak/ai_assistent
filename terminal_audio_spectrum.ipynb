```python
# Version 1
import plotext
import numpy as np
import pyaudio
import struct
import wave
import time

chunk = 4410
sample_format = pyaudio.paInt16
channels = 1
fs = 44100
seconds = 10
device_index = 2
filename = "audio_spectrum.wav"

print(f'\n... Recording {seconds} seconds of audio initialized ...\n')
print(
    f'\n\n>>> RECORDING PLEASE SPEAK CLEARLY <<<  >>> RECORDING PLEASE SPEAK CLEARLY <<<  >>> RECORDING PLEASE SPEAK CLEARLY <<<\n\n')

p = pyaudio.PyAudio()
stream = p.open(format=sample_format,
                channels=channels,
                rate=fs,
                input_device_index=device_index,
                frames_per_buffer=chunk,
                input=True,
                output=True,)

plotext.title("Spectrum")
plotext.clc()

x = np.arange(0, chunk, 1)

frames = []
start_time = time.time()

while time.time() - start_time < seconds:
    plotext.clt()
    plotext.cld()

    y = np.array(struct.unpack(str(2 * chunk) + 'B', stream.read(chunk)), dtype='b')[::2] + 128
    plotext.plot(x, y, color="green", marker="_", label="Spectrum")
    plotext.sleep(0.01)
    plotext.xlabel(f'Frequency (Hz) - {round(time.time() - start_time, 2)} seconds')
    plotext.ylim(0, 260)
    plotext.xlim(0, 4500)
    plotext.show()

    # Append audio frames to the list
    frames.append(y.tobytes())

# Stop and close the stream
stream.stop_stream()
stream.close()
p.terminate()

# Save the recorded data as a WAV file
wf = wave.open(filename, 'wb')
wf.setnchannels(channels)
wf.setsampwidth(p.get_sample_size(sample_format))
wf.setframerate(fs)
wf.writeframes(b''.join(frames))
wf.close()

plotext.clear_data()
```

```python
# Version 2
import plotext
import numpy as np
import pyaudio
import struct
import wave

chunk = 4410
sample_format = pyaudio.paInt16
channels = 1
fs = 44100
seconds = 10
device_index = 2
filename = "audio_spectrum.wav"

print(f'\n... Recording {seconds} seconds of audio initialized ...\n')
print(f'\n\n>>> RECORDING PLEASE SPEAK CLEARLY <<<  >>> RECORDING PLEASE SPEAK CLEARLY <<<  >>> RECORDING PLEASE SPEAK CLEARLY <<<\n\n')

p = pyaudio.PyAudio()
stream = p.open(format=sample_format,
                channels=channels,
                rate=fs,
                input_device_index=device_index,
                frames_per_buffer=chunk,
                input=True,
                output=True,)

plotext.title("Spectrum")
plotext.clc()

x = np.arange(0, chunk, 1)

frames = []
for i in range(fs):
    plotext.clt()
    plotext.cld()

    y = np.array(struct.unpack(str(2 * chunk) + 'B', stream.read(chunk)), dtype='b')[::2] + 128
    plotext.plot(x, y, color="green", marker="_", label="Spectrum")
    plotext.sleep(0.01)
    plotext.xlabel("Frequency")
    plotext.ylabel("Amplitude")
    plotext.ylim(0, 260)
    plotext.xlim(0, 4500)
    plotext.show()

    # Append audio frames to the list
    frames.append(y.tobytes())

# Stop and close the stream
stream.stop_stream()
stream.close()
p.terminate()

# Save the recorded data as a WAV file
wf = wave.open(filename, 'wb')
wf.setnchannels(channels)
wf.setsampwidth(p.get_sample_size(sample_format))
wf.setframerate(fs)
wf.writeframes(b''.join(frames))
wf.close()

plotext.clear_data()
```
