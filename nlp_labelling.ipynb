```python
#Needed Functions
from collections import Counter
import spacy
from spacy.lang.en.stop_words import STOP_WORDS

def lemmatize(doc):
    return [token.lemma_ for token in doc if not
           (token.is_punct or token.is_space or token.lower_ in STOP_WORDS)]

def tf(w, doc):
    c = 0
    for token in doc:
        if token.lemma_ == w:
            c = c + 1
    return c

def idf(w,docs):
    c = 0
    num_docs = len(docs)
    for doc in docs:
        if tf(w, doc) > 0:
            c = c + 1
    return c / num_docs
```
